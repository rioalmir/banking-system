```python
menu = '''
[d] Depositar
[s] Sacar
[e] Extrato
[q] Sair

=> '''


saldo = 0
limite = 500
extrato = ''
numero_saques = 0
limite_saques = 3

while True:

    opcao = input(menu)

    if opcao == 'd':
        print('Depósito')
        print('--------------------')
        valor_deposito = float(input('Qual valor deseja depositar? R$'))
        if valor_deposito > 0:
            confirmacao = str(input(f'O Valor depositado será R$ {valor_deposito} . Confirma? [s] ou [n]'))
            if confirmacao == 's':
                saldo += valor_deposito
                extrato = extrato + f'Depósito de R${valor_deposito:.2f}\n'
            else:
                print('Tente novamente...')
        else:
            print('Valor inválido, tente novamente')


    elif opcao == 's':
        print('Saque')
        print('--------------------')
        valor_saque = float(input('Qual valor deseja sacar? R$'))
        if valor_saque <= limite and valor_saque <= saldo and numero_saques < limite_saques  and valor_saque > 0:
            confirmacao = str(input(f'O Valor sacado será R$ {valor_saque} . Confirma? [s] ou [n]'))
            if confirmacao == 's':
                    saldo -= valor_saque
                    extrato += f'Saque de R${valor_saque:.2f}\n'
                    numero_saques += 1
            else:
                print('Comando inválido, tente novamente.')
        else:
            print(f'Você tem direito a 3 saques diários de até no máximo R$500 . Seu saldo atual é de {saldo} e você já fez {numero_saques} saques. Por favor tente novamente seguindo as condições informadas')
        

    elif opcao == 'e':
        print('Extrato')
        print('--------------------')
        print(extrato)
        print(f'Saldo: R${saldo} ')

    elif opcao == 'q':
        break

    else:
        print('Operação inválida, por favor selecione novamente a operação desejada.')
```
